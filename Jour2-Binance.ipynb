```python
import requests

r = requests.get("https://api.binance.com/api/v1/trades?symbol=ETHBTC")
data = r.json()
data
```

```python
somme = 0
for event in data:
    somme += float(event["price"])
    
moyenne = somme / len(data)
print(moyenne)
```

```python
from random import randint
# import random.randint

flux = 0
for event in data:
    if randint(0,1) == 0:
        flux += float(event["price"]) * float(event["qty"])
    
print(flux)
```
