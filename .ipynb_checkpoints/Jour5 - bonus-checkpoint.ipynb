```python
"""
    Décorateur 
"""

def mon_decorateur(f):
    print("Application du décorateur")
    def nouvelle_fonction(name):
        print("Je change le name")
        name = name.upper()
        return f(name)
    return nouvelle_fonction


@mon_decorateur
def dire_bonjour(name):
    print(f"Bonjour {name}")


dire_bonjour("Martin")

dire_bonjour("Lionel")
dire_bonjour("Leslie")
dire_bonjour("Christophe")
```

```python
from bs4 import BeautifulSoup
from textblob import TextBlob

r = requests.get("https://www.imdb.com/title/tt7286456/reviews?ref_=tt_urv")

bf = BeautifulSoup(r.text)
```

```python
liste = bf.select(".review-container .show-more__control")
liste_notes = bf.select(".review-container .rating-other-user-rating")
for review, note in zip(liste, liste_notes):
    blob = TextBlob(review.text)
    n = note.text
    polarite = (blob.sentiment.polarity + 1)*5
    print(f"note={n}, pol={polarite}")
    
```
